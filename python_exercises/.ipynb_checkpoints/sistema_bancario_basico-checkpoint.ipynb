# DOCUMENTAÇÃO 

Documentação do Programa Bancário
Visão Geral
O programa bancário é um sistema simples que simula algumas operações bancárias, como depósito, saque e consulta de extrato. Ele foi desenvolvido utilizando a linguagem de programação Python e pode ser executado em um ambiente de desenvolvimento Python, como o Jupyter Notebook.

O programa possui um menu com as seguintes opções:

Depósito: Permite ao usuário fazer um depósito na conta bancária.
Saque: Permite ao usuário fazer um saque da conta bancária.
Extrato: Mostra ao usuário o extrato bancário, contendo as transações realizadas.
Sair: Encerra a execução do programa.
O programa mantém o controle do saldo total da conta bancária, a quantidade de saques realizados e o histórico de transações (extrato). Além disso, possui uma função para limpar a tela do console e aguardar um curto intervalo de tempo para atualização.

Funções Principais
LimpaTela()
A função LimpaTela() é responsável por limpar a tela do console e aguardar um pequeno intervalo de tempo para que a tela seja atualizada.

deposito()
A função deposito() permite ao usuário fazer um depósito na conta bancária. Ela exibe o saldo atual da conta, solicita o valor a ser depositado e realiza as validações necessárias. Se o valor for válido, o saldo da conta é atualizado e a transação é registrada no extrato.

saque()
A função saque() permite ao usuário fazer um saque da conta bancária. Ela exibe uma mensagem de boas-vindas e solicita o valor a ser sacado. São realizadas validações para verificar se o valor é válido (não ultrapassa o saldo disponível, não é negativo e não ultrapassa o limite de saques permitidos). Se o valor for válido, o saldo da conta é atualizado, a quantidade de saques é incrementada e a transação é registrada no extrato.

extrato_bancario()
A função extrato_bancario() exibe o extrato bancário, mostrando todas as transações realizadas. As transações são exibidas em ordem numérica, juntamente com a descrição da transação.

Fluxo Principal
O programa começa exibindo uma mensagem de boas-vindas e o menu de opções. O usuário pode escolher uma das opções digitando o número correspondente. Após a escolha, o programa executa a função correspondente ou exibe uma mensagem de erro caso a opção seja inválida. Após a execução da opção escolhida, o menu é exibido novamente até que o usuário escolha a opção de sair.

O programa utiliza variáveis auxiliares para controlar o saldo total, a quantidade de saques e o extrato bancário. Essas variáveis são atualizadas durante as operações de depósito, saque e consulta de extrato.

Requisitos de Execução
Python 3.x instalado
Biblioteca IPython instalada
Executando o Programa
Certifique-se de que o Python 3.x está instalado em seu sistema.
Abra um ambiente de desenvolvimento Python, como o Jupyter Notebook.
Copie o código do programa para o ambiente de desenvolvimento.
Execute o código.
Siga as instruções exibidas no console para interagir com o programa.
Quando desejar sair, escolha a opção "4 - Sair".

PS: O objetivo principal desse programa é servir de teste para o uso de variáveis globais em escopos de funções, uma prática não recomendada, mas possível de se executar.

# CÓDIGO

```python
import platform
import os
from IPython.display import clear_output
import time

def limpatela():
    clear_output(wait=True)
    time.sleep(1)  # Aguarda um pequeno intervalo para que a tela seja atualizada


def deposito():
    
    global total_conta
    global extrato
    
    print('Seja muito bem vindo, estimado cliente!!')
    print(f"No momento atual, seu saldo é de R$ {total_conta:.2f}.")
    
    deposito = float(input("Quanto deseja depositar na sua conta? "))
    
    while deposito <= 0:
        deposito = float(input("Valor inválido! Quanto deseja depositar em sua conta? "))
        
    total_conta += deposito
    print(f"Seu novo saldo é de R$ {total_conta:.2f}.")
    extrato.append(f"Seu novo saldo é de R$ {total_conta:.2f}.")


def saque():
    
    global total_saque
    global total_conta
    global extrato
    
    
    print('Olá, caro cliente! Você escolheu a opção saque.')
    
    saque = float(input("Quanto deseja sacar? "))
    while (saque > total_conta) or (saque < 0) or (saque > 500):
        print("Valores inválidos para saque!")
        saque = float(input("Digite um valor adequado para saque: "))
        
    if saque == 0:
        print('Você escolheu não sacar, tenha um bom dia.')
    else:
        print(f'Você sacou R$ {saque:.2f}. Seu novo valor na conta é de R$ {total_conta - saque:.2f}.')
        total_saque += 1  
        print(f'Esse foi o seu {total_saque:.2f}º saque.')
        total_conta -= saque
        extrato.append(f'Você sacou R$ {saque:.2f}. Seu novo valor na conta é de R$ {total_conta - saque:.2f}.')
        extrato.append(f'Esse foi o seu {total_saque:.2f}º saque.')
    
def extrato_bancario():
    global extrato
    
    for indice, retorno in enumerate(extrato):
        print(f'{indice + 1}º - {retorno}')
    
def welcome():
    print('SEJA BEM VINDO AO BANCO DO LUIZ')

    print("""Há 3 opções disponíveis aqui hoje:
      [1] - Depósito
      [2] - Saque
      [3] - Extrato
      [4] - Sair
      """)

# Variáveis auxiliares

total_saque = total_conta = 0
extrato = auxiliar = list()

while True:
    limpatela()
    welcome()
    opcao = int(input("Escolha a sua opção: "))

    while opcao not in (1, 2, 3, 4):
        opcao = int(input("Escolha uma opção válida: "))

    if opcao == 1:
        deposito()
        limpatela()
    elif opcao == 2:
        if total_saque < 3:
            saque()
            limpatela()
        else:
            print('Você não tem mais saques disponíveis no momento.')
            limpatela()
    elif opcao == 3:
        if (total_saque == 0) and (total_conta == 0):
            print("Você ainda não tem um extrato bancário, realize transações e depois volte aqui")
        else:
            extrato_bancario()
    else:
        print("Você escolheu sair do programa.")
        break
```
